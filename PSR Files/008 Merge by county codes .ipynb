```python
# Required Dependencies

import pandas as pd
import numpy as np
import os
```

```python
# Files to Load

cal = "California_Final_100k.csv"
county = "County Codes.csv"
```

```python
# Read 'cal' and 'county' files

cal_final = pd.read_csv(cal)
county_final = pd.read_csv(county)
```

```python
# Merge 'cal_final' and 'county_final' files

cal_merge = pd.merge(cal_final, county_final, on="County", how="left")
cal_merge.set_index('County Code', inplace=True)
cal_merge
```

```python
cal_merge.shape
```

```python
# Save 'cal_merge' to csv

cal_merge.to_csv("California_Final_100K_County_Code.csv")
```
