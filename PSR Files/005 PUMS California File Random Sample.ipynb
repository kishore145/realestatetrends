```python
# Required Dependencies

import pandas as pd
import numpy as np
import os
```

```python
# Files to load

CalCsv = "California_1.csv"
```

```python
# Read file CalCsv

cal = pd.read_csv(CalCsv)
cal.shape
```

```python
# Print 'cal'

cal
```

```python
# Compile random sample 100,000 responses

calr100k = cal.sample(n=100000)
calr100k.shape
```

```python
# Print 'calr100k'

calr100k
```

```python
# Saved calr100k to a new csv file

calr100k.to_csv("California_1_Random_Sample_100K.csv")
```
