```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load 
tj_grocery = "TJ_Grocery.csv"
wh_grocery = "WH_Grocery.csv"
sw_grocery = "SW_Grocery.csv"
hospitals = "Hospitals_CA.csv"
```

```python
file_df1=pd.read_csv(tj_grocery)
file_df2=pd.read_csv(wh_grocery)
file_df3=pd.read_csv(sw_grocery)
file_df4=pd.read_csv(hospitals)
```

```python
df=pd.concat([file_df1[file_df1['State']=='CA'],file_df2[file_df2['State']=='CA'], file_df3[file_df3['State']=='CA'], file_df4])
    
file_df=df[[ 'Business Name', 'Category', 'City',  'State', 'County', 
       'County Code (FIPS)', 'County Subdivision Code (FIPS)',
       'Place Code (FIPS)']].copy()
file_df.head()
```

```python
new_df=file_df.groupby(['County','State','Category']).count()
```

```python
new_df=new_df.reset_index().sort_values(['State','County','Category'])
```

```python
new_df.head()
```

```python
#find long,lat from County
from geopy import geocoders  
gn = geocoders.GeoNames(username='vsaxena.ubc')
```

```python
grocery_df=new_df[new_df['Category']=='grocery']
hospital_df=new_df[new_df['Category']=='hospital']
```

```python
#create city/combined list
city=[]
for c in new_df['City']:
    city.append(c)

lon=[]
lat=[]
county=[]
for row in new_df['County']:
    #print(row)
    if ('***' in row):
        print("Skiping")
    else:
        gl=gn.geocode(row +", CA")
        lon.append(gl.longitude)
        lat.append(gl.latitude)
        county.append(row)
        print (f"{gl.longitude}   {gl.latitude}")
```

```python
#create city/grocery list
city_g=[]
for c in grocery_df['City']:
    city_g.append(c)
    
#create city/hospital list
city_h=[]
for c in hospital_df['City']:
    city_h.append(c)    
```

```python
lon_g=[]
lat_g=[]
county_g=[]
for row in grocery_df['County']:
    #print(row)
    if ('***' in row):
        print("Skiping")
    else:
        gl=gn.geocode(row +", CA")
        lon_g.append(gl.longitude)
        lat_g.append(gl.latitude)
        county_g.append(row)
        print (f"{gl.longitude}   {gl.latitude}")
        
```

```python
lon_h=[]
lat_h=[]
county_h=[]
for row in hospital_df['County']:
    #print(row)
    if ('***' in row):
        print("Skiping")
    else:
        gl=gn.geocode(row +", CA")
        lon_h.append(gl.longitude)
        lat_h.append(gl.latitude)
        county_h.append(row)
        print (f"{gl.longitude}   {gl.latitude}")
```

```python
df1=pd.DataFrame({"County":county, "Longitude":lon, "Latitude":lat})
df2=new_df.merge(df1, on="County", how="left")
df2.head()
```

```python
df2.to_csv("Amenties.csv")
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# 1. Draw the map background
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=37.5, lon_0=-119,
            width=1E6, height=1.2E6)
m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')


m.scatter(lon_g, lat_g, latlon=True,
          c='g', s=city_g*5,
          cmap='Reds', alpha=0.5,label="Grocery")

m.scatter(lon_h, lat_h, latlon=True,
          c='r', s=city_h*5,
          cmap='Reds', alpha=0.5,label="Hospitals")

# make legend with dummy points

plt.legend()
```
